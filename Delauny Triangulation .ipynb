```python
a = np.array([[1,2,3],[2,4,5], [7,8,9], [10,11,22]]) 
b= np.array([[3,4,5],[7,8,9],[10,11,22], [1,2,3]])
threshold = 0.5
a_rep = np.repeat(a,b.shape[0],axis=0)
b_rep = np.tile(b,(a.shape[0],1))
diff = np.linalg.norm(a_rep -b_rep,axis=1)

bool_mask = np.any(np.reshape(diff<threshold, (-1,b.shape[0])),axis = 1) 
repeatbility = float(np.sum(bool_mask)) / float(a.shape[0])
```
